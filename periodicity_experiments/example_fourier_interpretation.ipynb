```python
import numpy as np
from scipy.fftpack import fft
import matplotlib.pyplot as plt
```

```python
# Time period
t = np.arange(0, 10, 0.01);

# Create a sine wave with multiple frequencies(1 Hz, 2 Hz and 4 Hz)
a = np.sin(2*np.pi*t) + np.sin(2*2*np.pi*t) + np.sin(4*2*np.pi*t);

plt.plot(t,a)
```

```python
# Number of sample points
N = 600
# sample spacing
T = 0.75 / 600.0
x = np.linspace(0.0, N*T, N)
y = np.sin(10.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
yf = fft(y)
#for N%2=0 first N/2 elements is relative to positive frequencies
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)
plt.figure(1)
plt.subplot(211)
plt.plot(x,y)
plt.grid()

plt.subplot(212)
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.grid()
plt.show()
```
