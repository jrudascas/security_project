## Method
- Select two scenarios when convergence results change dramatically (convergence successful for a trainning dataset but fails when one day is incremented on trainning dataset). Scenarios selection was made based on experiment_03_1_SEPPexp_convergence results.
- Get omega and theta parameters on each scenario.
- Plot parameters values along iterations.


## Hypothesis:
Omega value determines the convergence of the model. In the convergence scenario omega variations are small when iterations are incremented, while in the non-convergence scenarios the variations become very large.
